# Quick Test with mushrooms

This notebook uses ForestSim to compute similarities, and plot a 2-D multidimensional projection with the result.

Steps to complete:
 - Compare with PCA
 - run a simple clustering and describe the results.

```python
from sklearn import preprocessing
import pandas as pd


data = pd.read_csv("../data/mushrooms.csv", sep=",") #.sample(n=2000, random_state=0)
y, X_raw = data.iloc[:,0], data.iloc[:,1:]
X_raw.shape

OHE = preprocessing.OneHotEncoder()
X = OHE.fit_transform(X_raw)
X = pd.DataFrame(data = X.toarray(), columns = OHE.get_feature_names(), index = X_raw.index )
X.head(2), X.shape

OHE = preprocessing.OneHotEncoder()
y_ = OHE.fit_transform(y.values.reshape(-1, 1))
y_ = pd.DataFrame(data = y_.toarray(), columns = OHE.get_feature_names(), index = y.index )
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=5, n_jobs=-1, max_depth=2, min_samples_split=300)
FS = ForestSim(clf)
FS.fit(X,y=y.values, nb_repet=6)
```

```python
FS.apply_MDS(n_instance = 400)
FS.project_MDS_2D()
```
